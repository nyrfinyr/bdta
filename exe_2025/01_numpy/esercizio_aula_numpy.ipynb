```python
import numpy as np
```

**1. Creazione dell'array**

Definisci una funzione che crea un array NumPy di due dimensioni (30 * 24) che riporta le temperature orarie del mese di Settembre.

Inizializza i valori con numeri decimali casuali compresi tra 10 e 30 gradi Celsius (tip: usa *np.random.uniform()*).

Note: for reproducibility, set seed at the beginning of the function definition using *np.random.seed(100)*



```python
np.random.seed(100)
temperatures = np.random.uniform(10,30,size=(30,24))
```

**2. Temperatura più alta e più bassa**

Trova i valori di temperatura più alta e più bassa

```python
max_value = np.max(temperatures)
min_value = np.min(temperatures)
print('max temperature found: ',max_value)
print('min temperature found: ',min_value)
```

**3. Temperatura media giornaliera**

Per ogni giorno (quindi per ogni riga), calcola la temperatura media e crea un nuovo array di 30 elementi che rappresenta la temperatura media giornaliera.

Trova i giorni con la temperatura media più alta e la temperatura media più bassa.

```python
mean_day_temperature = np.mean(temperatures, axis=1)
mean_size = mean_day_temperature.size
print(mean_day_temperature)
print('mean size: ',mean_size)
```

**4. Aumento di temperatura**

Sono stati rilevati alcuni problemi nella calibrazione del misuratore, è necessario aumentare di 2 gradi ogni temperatura rispetto a quanto misurato.

```python
temperature_increased = temperatures + 2
print(temperature_increased)
```

**5. Modifica di alcuni valori**

Modifica i valori dell'array in modo che il valore massimo possibile sia 30.

```python
temperature_increased[temperature_increased > 30] = 30
temperature_increased.apply(lambda x: 30 if x > 30 else x)
```

**6. Quante volte sono stati registrati più di 23 gradi?**

```python
count_greater_23 = np.count_nonzero(temperature_increased > 23)
print(count_greater_23)
```

**7. Confronto di valori medi**

Calcola la differenza tra la temperatura media dei primi 3 giorni di Settembre e quella degli ultimi 3 giorni.

```python
temperature_increased_daily_mean = np.mean(temperature_increased, axis=1)
first_3_days_mean = np.mean(temperature_increased_daily_mean[:3])
last_3_days_mean = np.mean(temperature_increased_daily_mean[-3:])
diff = np.abs(first_3_days_mean - last_3_days_mean)
print(diff)
```

**8. Creazione di una nuova colonna**

Crea una nuova colonna con il valore della mediana giornaliera

```python
print('shape:',temperature_increased.shape)
daily_median = np.median(temperature_increased, axis=1)
temperature_increased_median = np.column_stack((temperature_increased, daily_median))
print('shape:',temperature_increased_median.shape)
print(temperature_increased_median)
```
