1. Inserisci in un dataframe il contenuto del file "dipendenti.csv".

   Quante righe e quante colonne ci sono? Ci sono duplicati?

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2025/pandas2/data/dipendenti.csv', sep=';')
df.count().max()
df.duplicated().value_counts()
```

2. Crea una nuova colonna chiamata **Livello Esperienza**, utilizzando la colonna **Anni di esperienza**. Se l'esperienza è minore di 5 anni, mappa il valore a "Junior", tra 5 e 10 a "Mid" e più di 10 a "Senior".

```python
df['Livello Esperienza'] = df['Anni di esperienza'].map(lambda anni : 'Junior' if anni < 5 else('Senior' if anni > 10 else 'Mid'))
df.head()
```

3. Sostituisci i nomi dei dipartimenti con le seguenti abbreviazioni: Vendite -> 'VEN', Information Technology -> 'IT', Human Resources -> 'HR'. Quanti dipendenti ci sono per ogni dipartimento?

```python
df['Dipartimento'] = df['Dipartimento'].map(lambda dip : 'VEN' if dip == 'Vendite' else('IT' if dip == 'Information Technology' else 'HR'))
df.head()
```

4. Crea un'altra colonna con i valori di età discretizzati in 5 gruppi, a quale gruppo di età appartiene la maggior parte dei dipendenti?

```python
df['Età discretizzata'] = pd.cut(df['Età'], 5)
df['Età discretizzata'].value_counts()
```

5. Sostituisci i valori mancanti della colonna età con il valore medio di età per i dipendenti dello stesso settore.

```python
mean_transform_series = df.groupby('Dipartimento')['Età'].transform('mean')
df['Età'] = df['Età'].fillna(mean_transform_series)
df.head()
```
