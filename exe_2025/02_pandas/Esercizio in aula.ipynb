1. Crea il DataFrame **vendite** che contiene la quantità di torte, biscotti e taralli (colonne) nei diversi giorni della settimana (righe).

  Inserisci valori casuali compresi tra 5 e 50 (np.random.seed(100)).

```python
np.random.seed(100)
vendite = {'torte': np.random.randint(5, 50, 7), 'biscotti':  np.random.randint(5, 50, 7), 'taralli':  np.random.randint(5, 50, 7)}
vendite = pd.DataFrame(data = vendite)
vendite.index = ['lun', 'mar', 'mer', 'gio', 'ven', 'sab', 'dom']
vendite
```
