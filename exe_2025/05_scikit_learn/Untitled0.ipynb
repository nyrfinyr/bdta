## Scikit Learn!!

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import sklearn as sk
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

### KNN Classifier
Nota che stiamo calcolando l'accuracy sugli stessi dati di training: molto probabilmente stiamo overfittando fortissimo

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, Y)
knn.predict(X)

accuracy_score(Y, knn.predict(X))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X, Y)
dtc.predict(X)

accuracy_score(Y, dtc.predict(X))
```

### Train, Test, Split
stavolta non dovremmo overfittare

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, Y_train)
accuracy_score(Y_test, knn.predict(X_test))
```

### Vediamo uno scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

dtc = DecisionTreeClassifier()
dtc.fit(X_train_scaled, Y_train)
accuracy_score(Y_test, dtc.predict(scaler.transform(X_test)))
```
