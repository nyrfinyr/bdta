**Analisi dei dati anagrafici dei dipendenti di un'azienda**

```python
import pandas as pd
import numpy as np
```

1. Inserisci in un dataframe il contenuto del file "dipendenti.csv".

   Quante righe e quante colonne ci sono?

```python
df = pd.read_csv('../data/dipendenti.csv', sep=';')
n_columns = len(df.columns)
n_rows = len(df.index)
print(f"rows: {n_rows} columns: {n_columns}")
```

2. Crea una nuova colonna chiamata **Livello Esperienza**, utilizzando la colonna **Anni di esperienza**. Se l'esperienza è minore di 5 anni, mappa il valore a "Junior", tra 5 e 10 a "Mid" e più di 10 a "Senior".

```python
df['Livello Esperienza'] = pd.cut(df['Anni di esperienza'], (0, 5, 10, np.inf), labels=['Junior', 'Mid', 'Senior'])
df.head()
```

3. Sostituisci i nomi dei dipartimenti con le seguenti abbreviazioni: Vendite -> 'VEN', Information Technology -> 'IT', Human Resources -> 'HR'.

```python
abbr_map = {'Vendite': 'VEN', 'Information Technology': 'IT', 'Human Resources': 'HR'}
df['Dipartimento'] = df['Dipartimento'].map(abbr_map)
df.head()
```

4. Crea un'altra colonna con i valori di età discretizzati in 5 gruppi, a quale gruppo di età appartiene la maggior parte dei dipendenti?

```python
df['Età bin'] = pd.cut(df['Età'], bins=5)
df['Età bin'].value_counts()
```

5. Sostituisci i valori mancanti della colonna età con il valore medio di età per i dipendenti dello stesso settore.

```python
mean_series = df.groupby(['Dipartimento'])['Età'].transform('mean')
df['Età'] = df['Età'].fillna(mean_series)
df.head()
```

```python
mean_s = df.groupby(['Dipartimento'])['Età'].transform(lambda g : g.mean())
```

6. Verifica se, in media, lo stipendio di chi vive a Milano è superiore rispetto a quello di chi vive a Roma.

```python
city_dict = df.groupby(['Città'])['Stipendio'].mean().to_dict()
city_dict['Milano'] > city_dict['Roma']
```
