```python
import pandas as pd
import numpy as np
```

# Grouping data

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

## Which continent drinks more beer on average?

```python
drinks.groupby(['continent'])['beer_servings'].mean().idxmax()
```

## For each continent print the statistics for wine consumption

```python
drinks.groupby(['continent'])['wine_servings'].describe()
```

```python
drinks.groupby(['continent'])['wine_servings'].agg(['mean', 'std', 'min', 'max'])
```

## Print the mean alcohol consumption per continent for every column

```python
drinks.groupby(['continent']).mean(numeric_only=True)
```

## Print the median alcohol consumption per continent for every column

```python
drinks.groupby(['continent']).median(numeric_only=True)
```

## Print the mean, min and max values for spirit consumption

```python
drinks.groupby(['continent']).agg('mean', 'min', 'max')
```

# Pivoting

```python
df = pd.read_csv('../data/titanic.csv')
df.head()
```

## Find average survival rate by gender

```python
df.groupby('sex')['alive'].apply(lambda x : (x == 'yes').mean())

df_tmp = df.copy()
df_tmp['survived_numeric'] = df_tmp['alive'].map({'yes' : 1, 'no': 0})
df_tmp.groupby('sex')['survived_numeric'].mean()
```

## Create a Pivot table and find survival rate by gender on various classes

```python
df_tmp = df.copy()
df_tmp['survived_numeric'] = df_tmp['alive'].map({'yes' : 1, 'no': 0})
pd.pivot_table(data=df_tmp, index='sex', columns='pclass', values='survived_numeric', aggfunc='mean')
```

## Create a Pivot table and find survival rate by gender, age wise of various classes

```python
pd.pivot_table(data=df_tmp, index=['sex', 'age'], columns='pclass', values='survived_numeric')
```

## Partition each of the passengers into four categories based on their age (age categories (0, 10), (10, 30), (30, 60), (60, 80)) and then create a pivot table that counts survival by gender, categories wise age of various classes

```python
# pd.cut()
# pivot_table()
```

```python
bins = [0, 10, 30, 60, 80]
df['age_bins'] = pd.cut(df['age'], bins)
df['survived_numeric'] = df['alive'].map({'yes': 1, 'no': 0})
df.pivot_table(index = ['sex', 'age_bins'], columns='pclass', values='survived_numeric', aggfunc='mean')
```
