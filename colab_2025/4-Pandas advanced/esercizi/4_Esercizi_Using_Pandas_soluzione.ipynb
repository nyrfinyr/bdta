```python
import pandas as pd
import numpy as np
```

# Load data

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url, header=None, names=['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'])

iris.head()
```

# Data cleaning

## Missing values

### Is there any missing value in the dataframe?

```python
pd.isnull(iris).sum()
```

### Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris.iloc[10:30,2:3] = np.nan
iris.head(30)
```

### Which column has the maximum number of missing values?

```python
missing_per_columns = pd.isnull(iris).sum()
missing_per_columns.index[missing_per_columns.argmax()]
```

```python
missing_per_columns.idxmax()
```

### Try to substitute the NaN values with two methods:
- replace null values with column mean (apply it to a copy of the dataframe)
- replace null values with 1.0



```python
new_iris = iris.copy()
new_iris['petal_length'] = new_iris['petal_length'].fillna(new_iris['petal_length'].mean())
new_iris.head(20)
```

```python
iris.petal_length = iris.petal_length.fillna(1)
iris.head(20)
```

### Set the first 3 rows as NaN

```python
iris.iloc[0:3 ,:] = np.nan
iris.head()
```

### Delete the rows that have all NaN

```python
iris = iris.dropna(how='all')
iris.head(20)
```

### Reset the index so it begins with 0 again

```python
iris = iris.reset_index(drop = True)
iris.head(20)
```

## Duplicates

### Does the dataframe contain duplicated rows? If any, visualize all duplicated rows (don't omit first or last occurrences)

```python
iris_duplicate_rows = iris[iris.duplicated(keep=False)]
iris_duplicate_rows
```

### Which row is the most repeated?

```python
iris_duplicate_rows.groupby(list(iris_duplicate_rows.columns.values)).size()
```

### Drop duplicated rows

```python
iris = iris.drop_duplicates()
```

## Detect outliers

```python
iris_copy = iris.loc[:,"sepal_length":"petal_width"].copy()
a = iris_copy[(iris_copy > iris_copy.quantile(.85)).all(axis=1)]
b = iris_copy[(iris_copy < iris_copy.quantile(.25)).all(axis=1)]
```

```python
iris_copy.drop(index=a.index).drop(index=b.index)
```

# Data transformation

## Replace class values by removing "Iris-" prefix (use a dictionary)

```python
class_map_values = {}
for class_val in iris['class'].unique():
  class_map_values[class_val] = class_val.replace("Iris-", "")
class_map_values
```

```python
iris['class'] = iris['class'].replace(class_map_values)
iris
```

## Delete columns
Delete for example class column

```python
del iris['class']
iris
```

## How to normalize all columns in a dataframe?
- Normalize all columns of df by subtracting the column mean and divide by standard deviation.
- Range all columns of df such that the minimum value in each column is 0 and max is 1.

```python
# Solution Q1
out1 = iris.apply(lambda x: ((x - x.mean())/x.std()).round(2))
print('Solution Q1\n',out1)

# Solution Q2
out2 = iris.apply(lambda x: ((x.max() - x)/(x.max() - x.min())).round(2))
print('Solution Q2\n', out2)
```

## Merging and concatenating dataframes

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])
```

### Concatenate the first two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1, data2])
all_data
```

### Concatenate the first two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.concat([data1, data2], axis = 1)
all_data_col
```

### Merge all_data and data3 along the subject_id value

```python
pd.merge(all_data, data3, on='subject_id')
```

### Merge data1 and data2 on the basis of the same 'subject_id'

```python
pd.merge(data1, data2, on='subject_id', how='inner')
```

### Merge data1 and data2 on the basis of the same 'subject_id' and mantaining also the rows that have a subject id that doesn't exist in the other dataframe

```python
pd.merge(data1, data2, on='subject_id', how='outer')
```

### Rename column 'subject_id' from data2 with 'sid' and re-apply the previous merge operation

```python
new_data2 = data2.rename(columns={'subject_id': 'sid'})
new_data2.columns
```

```python
pd.merge(data1, new_data2, left_on='subject_id', right_on='sid', how='outer')
```

### Create two new columns, one in data1 containing the numbers from 0 to 5 and one in data2 containing the numbers from 0 to 20 with step 4. Then merge (inner) data1 and data2 on subject_id and the new generated columns

```python
new_data1 = data1.copy()
new_data1['col'] = range(5)
print(new_data1)

new_data2 = data2.copy()
new_data2['col'] = range(0, 20, 4)
print(new_data2)
```

```python
match = pd.merge(new_data1, new_data2, on=['subject_id', 'col'])
match
```

# Grouping data

## Which continent drinks more beer on average?

```python
drinks.groupby('continent').beer_servings.mean()
```

## For each continent print the statistics for wine consumption

```python
drinks.groupby('continent').wine_servings.describe()
```

## Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continent').mean(numeric_only=True)
```

## Print the median alcohol consumption per continent for every column

```python
drinks.groupby('continent').median(numeric_only=True)
```

## Print the mean, min and max values for spirit consumption

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

# Pivoting

```python
df = pd.read_csv('titanic.csv')
df.head()
```

## Find average survival rate by gender

```python
result=df.groupby('sex')[['survived']].mean()
result
```

## Create a Pivot table and find survival rate by gender on various classes

```python
result = df.pivot_table('survived', index='sex', columns='class', aggfunc=np.mean)
result
```

## Create a Pivot table and find survival rate by gender, age wise of various classes

```python
result  =  df.pivot_table('survived', index=['sex','age'], columns='class')
result
```

## Partition each of the passengers into four categories based on their age (age categories (0, 10), (10, 30), (30, 60), (60, 80)) and then create a pivot table that counts survival by gender, categories wise age of various classes

```python
age = pd.cut(df['age'], [0, 10, 30, 60, 80])
result = df.pivot_table('survived', index=['sex',age], columns='pclass', aggfunc='count')
result
```
