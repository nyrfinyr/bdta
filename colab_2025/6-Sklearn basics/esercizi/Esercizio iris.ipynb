1. Creare un dataframe che contenga i valori di iris

```python
import pandas as pd 
import numpy as np
from sklearn import datasets 
```

```python
iris_bunch = datasets.load_iris(as_frame=True)
df:pd.DataFrame = iris_bunch.frame
df.head()
```

2. Creare due nuovi attributi che rappresentano l'area del petalo e del sepalo. Calcolare l'accuratezza con una divisione train /test 0.75/0.25 e un Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df['petal area'] = df['petal length (cm)'] * df['petal width (cm)']
df['sepal area'] = df['sepal length (cm)'] * df['sepal width (cm)']

X = df.drop(columns=['target'])
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, test_size=.25)
tree_classifier = DecisionTreeClassifier().fit(X_train, y_train)

y_predicted = tree_classifier.predict(X_test)
accuracy_score(y_test, y_predicted)
```

3. Creare una discretizzazione dei valori in 4 bin (per ogni colonna del dataset), sostituire al dataset iniziale e calcolare l'accuratezza

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target', 'petal_area', 'sepal_area']

kbd = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')

X = df.loc[:, 'sepal_length':'petal_width']
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, test_size=.25, random_state=42)

fit_kbd = kbd.fit(X_train)

X_train_discretized = fit_kbd.transform(X_train)
X_test_discretized = fit_kbd.transform(X_test)

tree_classifier = DecisionTreeClassifier(random_state=42).fit(X_train_discretized, y_train)

y_predicted = tree_classifier.predict(X_test_discretized)
accuracy_score(y_test, y_predicted)
```

4. Standardizzare il dataset del punto 2 e calcolare l'accuratezza con il classificatore di prima

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(columns=['target'])
y = df['target']

### Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, test_size=.25, random_state=42)

### Fitting e Transform dello Standard Scaler
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

### Decision Tree training
tree_classifier = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)

### Accuracy
y_predicted = tree_classifier.predict(X_test)
accuracy_score(y_test, y_predicted)
```

5. Applicare il logaritmo (trasformazione custom) a tutte le colonne del dataset del punto 2 e calcolare l'accuratezza con il classificatore di prima

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
X = df.drop(columns=['target'])
y = df['target']

### Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, test_size=.25, random_state=42)

### Custom transformation: log
#X = np.log(X)
log_transformer = FunctionTransformer(np.log1p, validate=True)
log_transformer.fit(X_train)

X_train = log_transformer.transform(X_train)
X_test = log_transformer.transform(X_test)

### Decision Tree training
tree_classifier = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)

### Accuracy
y_predicted = tree_classifier.predict(X_test)
accuracy_score(y_test, y_predicted)
```

6. Creare una nuova colonna che abbia valore 1 se e solo se l'area del petalo e l'area del sepalo hanno entrambe valore superiore a 4 e calcolare l'accuratezza con il classificatore di prima

```python
df.head()
```

```python
df['greater_than_4'] = (df['petal_area'] > 4) & (df['sepal_area'] > 4)

X = df.drop(columns=['target', 'petal_area', 'sepal_area'])
y = df['target']

### Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, test_size=.25, random_state=42)

### Decision Tree training
tree_classifier = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)

### Accuracy
y_predicted = tree_classifier.predict(X_test)
accuracy_score(y_test, y_predicted)
```
