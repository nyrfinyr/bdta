```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn import datasets
```

```python
def accuracy(pred, result):
  return 1-(abs(pred-result)).sum()/len(result)
```

```python
iris = datasets.load_iris()
```

```python
iris['target_names']
```

```python
iris['feature_names']
```

```python
X = iris.data
y = iris.target
```

```python
df = pd.DataFrame(X, columns=iris['feature_names'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
result = model.predict(X_test)
```

```python
accuracy(y_test, result)
```

```python
accuracy(y_train, model.predict(X_train) )
```

## Creare due nuovi attributi che rappresentano l'area del petalo e del sepalo

```python
df['areaSep'] = df['sepal length (cm)'] *df['sepal width (cm)']
df['areaPet'] = df['petal length (cm)'] *df['petal width (cm)']
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25)
```

```python
model.fit(X_train, y_train)
result = model.predict(X_test)
accuracy(y_test, result)
```

```python
accuracy(y_train, model.predict(X_train))
```

## Creare una nuova colonna che abbia valore 1 se e solo se l'area del petalo e l'area del sepalo hanno entrambe valore superiore a 4 e calcolare l'accuratezza con il classificatore di prima

```python
df.head()
```

```python
def checkArea(x,val=4):
  mask0 = x[:,0]>val
  mask1 = x[:,1]>val
  return mask0 & mask1
```

```python
checkArea(X)
```

```python
transformer = FunctionTransformer(checkArea, validate=True)
```

```python
boolArea = transformer.transform(df.loc[:,['areaSep','areaPet']].values)
```

```python
boolArea
```

```python
df["boolArea"] = boolArea
```

```python
df.head()
```

```python
df.tail()
```
