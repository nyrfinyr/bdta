```python
import pandas as pd 
import numpy as np
```

1. Crea il DataFrame **vendite** che contiene la quantità di torte, biscotti e taralli (colonne) nei diversi giorni della settimana (righe).

  Inserisci valori casuali compresi tra 5 e 50 (np.random.seed(100)).

```python
np.random.seed(100)
data = np.random.randint(5, 50, size=(7,3))
vendite = pd.DataFrame(data, columns=['torte', 'biscotti', 'taralli'], index=['LUN', 'MAR', 'MER', 'GIO', 'VEN', 'SAB', 'DOM'])
vendite.head()
```

2. Visualizza le vendite di torte durante tutta la settimana.

```python
vendite
```

3. Ottieni le vendite del giovedi.

```python
vendite.loc['GIO']
```

4. Visualizza le vendite di biscotti e taralli dal mercoledì alla domenica.

```python
vendite.loc['MER':, ['biscotti', 'taralli']]
```

5. Qual è il totale delle vendite per ogni prodotto durante la settimana?

```python
vendite.sum()
```

6. Qual è il giorno con il massimo delle vendite di torte?

```python
max_torte = vendite['torte'].max()
vendite[vendite['torte'] == max_torte]
vendite['torte'].idxmax()
```

7. Aggiungi la colonna **Totale Vendite Giorno** che rappresenta la somma delle vendite di tutti i prodotti per ciascun giorno.

```python
vendite['Totale Vendite Giorno'] = vendite.sum(axis=1).astype(np.int64)
vendite
```

8. Quali giorni hanno avuto un totale di vendite giornaliere superiori a 80?

```python
vendite[vendite['Totale Vendite Giorno'] > 80]
```

```python
vendite.where(lambda df : df['Totale Vendite Giorno'] > 80)
```
